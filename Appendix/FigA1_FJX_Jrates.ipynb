```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
atmlayers_sol=pd.read_csv('../output/out_atmlayers_sol.dat', skiprows=1, delim_whitespace=True)
jvalues_pcb_nocor = pd.read_csv('../output/pcb_nocorr.text', skiprows=377, delim_whitespace=True)
jvalues_solar_nocor = pd.read_csv('../output/solar_nocorr.text', skiprows=377, delim_whitespace=True)
jvalues_pcb_muscles= pd.read_csv('../output/pcb_muscles.text', skiprows=377, delim_whitespace=True)
jvalues_pcb_yates = pd.read_csv('../output/pcb_yates.text', skiprows=377, delim_whitespace=True)
```

```python
plt.figure(figsize=(11,7))
plt.subplot(1,3,1)
plt.plot(np.log10(jvalues_solar_nocor['O2']), atmlayers_sol['p'][1:]*100, color='blue', label='Earth')
plt.plot(np.log10(jvalues_pcb_nocor['O2']), atmlayers_sol['p'][1:]*100, ls='-', color='darkorange', label='PCb, BT-Settl')
plt.plot(np.log10(jvalues_pcb_yates['O2']), atmlayers_sol['p'][1:]*100, ls='--', color='darkorange', label='PCb, Yates+20')
plt.plot(np.log10(jvalues_pcb_muscles['O2']), atmlayers_sol['p'][1:]*100, color='red', label='PCb, MUSCLES')
plt.ylabel('P(Pa)', fontsize=20)
plt.xticks(fontsize=20)
plt.gca().invert_yaxis()
plt.yscale('log')
plt.yticks(fontsize=20)
plt.title('O2', fontsize=24)
plt.xlim(-16,-8)

plt.subplot(1,3,2)
plt.plot(np.log10(jvalues_solar_nocor['O3']), atmlayers_sol['p'][1:]*100, color='blue', label='Earth')
plt.plot(np.log10(jvalues_pcb_nocor['O3']), atmlayers_sol['p'][1:]*100, ls='-', color='darkorange', label='PCb, BT-Settl')
plt.plot(np.log10(jvalues_pcb_yates['O3']), atmlayers_sol['p'][1:]*100, ls='--', color='darkorange', label='PCb, Yates+20')
plt.plot(np.log10(jvalues_pcb_muscles['O3']), atmlayers_sol['p'][1:]*100, color='red', ls='-', label='PCb, MUSCLES')
plt.xlabel('log$_{10}$(J)', fontsize=20)
plt.xticks(fontsize=20)
plt.gca().invert_yaxis()
plt.yscale('log')
plt.yticks([])
plt.title('O3(O3P)', fontsize=24)
plt.xlim(-5,-1)

plt.subplot(1,3,3)
plt.plot(np.log10(jvalues_solar_nocor['O3(1D)']), atmlayers_sol['p'][1:]*100, color='blue', label='Earth')
plt.plot(np.log10(jvalues_pcb_nocor['O3(1D)']), atmlayers_sol['p'][1:]*100, ls='-', color='darkorange', label='PCb, BT-Settl')
plt.plot(np.log10(jvalues_pcb_yates['O3(1D)']), atmlayers_sol['p'][1:]*100, ls='--', color='darkorange', label='PCb, Yates+20')
plt.plot(np.log10(jvalues_pcb_muscles['O3(1D)']), atmlayers_sol['p'][1:]*100, color='red', ls='-', label='PCb, MUSCLES')
plt.xlim(-9,-2)
plt.xticks(fontsize=20)
plt.gca().invert_yaxis()
plt.yscale('log')
plt.yticks([])
plt.title('O3(O1D)', fontsize=24)
plt.legend(fontsize=20, bbox_to_anchor=(1.01, 0.6))
plt.show()
```
