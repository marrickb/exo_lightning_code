```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing spectra computed by NASA Planetary Spectrum Generator

see https://psg.gsfc.nasa.gov/index.php

```python
pcb_termspecnonox=pd.read_csv('../PSG/Full_term/pcb_term_musc_nonox.txt', skiprows=11, delim_whitespace=True)
pcb_termspec=pd.read_csv('../PSG/Full_term/pcb_term_musc.txt', skiprows=11, delim_whitespace=True)
```

# Results in Figure 11 from Braam et al. (2022)

```python
plt.figure(figsize=(8,6))
plt.plot(pcb_termspecnonox['Wave/freq'], pcb_termspecnonox['Total'],label='Base')
plt.plot(pcb_termspec['Wave/freq'], pcb_termspec['Total'],label='Li-NO$_\mathrm{y}$')
plt.legend(loc='best', fontsize=18)
plt.xlabel('Wavelength($\mu$m)', fontsize=18)
plt.ylabel('(R$_p$/R$_*$)$^2$', fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlim(0.8,10)
#plt.xscale('log')
plt.ylim(0.005414,0.00544)
#plt.savefig('linoxspectra.pdf', bbox_inches = 'tight', pad_inches=0)
plt.show()
```
